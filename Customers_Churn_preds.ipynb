```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv("customer_churn.csv")
```

```python
df.head()
```

```python
df.drop('customerID',axis=1,inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'].str.replace(' ', ''),errors='coerce')
```

```python
df['TotalCharges'].isna().sum()
```

```python
df.dropna(subset=['TotalCharges'], inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
No=df[df.Churn=='No'].tenure
yes=df[df.Churn=='Yes'].tenure
```

```python
plt.hist([yes,No],label = ['Churn=Yes','Churn=No'])
plt.xlabel('Tenure')
plt.ylabel('No.of customers')
plt.legend()
```

```python
No=df[df.Churn=='No'].MonthlyCharges
yes=df[df.Churn=='Yes'].MonthlyCharges

plt.hist([yes,No],label = ['Churn=Yes','Churn=No'])
plt.xlabel('Monthly Charges')
plt.ylabel('No.of customers')
plt.legend()
```

```python
def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes=='object':
            print(f'{column}:{df[column].unique()}')
```

```python
print_unique_col_values(df)
```

```python
df.replace('No internet service','No',inplace=True)
df.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df)
```

```python
cols_to_encode = ['gender','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV',
                  'StreamingMovies','PaperlessBilling','Churn']
for cols in cols_to_encode:
    df[cols].replace({'Yes':1,'No':0},inplace=True)
    
```

```python
df['gender'].replace({'Male':1,'Female':0},inplace=True)
```

```python
df.dtypes
```

```python
df=pd.get_dummies(data=df, columns=['InternetService'], dtype=int)
```

```python
df=pd.get_dummies(data=df, columns=['Contract'], dtype=int)
df=pd.get_dummies(data=df, columns=['PaymentMethod'], dtype=int)
```

```python
df.info()
```

```python
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python
df.sample(5)
```

```python
for column in df:
     print(f'{column}:{df[column].unique()}')
```

```python
df2=df.corr()
```

```python
plt.figure(figsize=(24,20))
sns.heatmap(data=df2,annot=True)
```

```python
X = df.drop('Churn',axis=1)
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state = 5)
```

```python
import tensorflow as tf
from tensorflow import keras 

model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,),activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(X_train,y_train, epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
predict  = model.predict(X_test)
predict[:5]
```

```python
y_test[:10]
```

```python
y_pred = []
for elem in predict:
    if elem>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]#compare with the y_test above and see how accurate your model is predicting!
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
